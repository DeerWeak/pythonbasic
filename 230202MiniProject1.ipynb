```python
aed_df_office = aed_df["설치기관명"]
list_aed_df_office = list(aed_df_office)
list_aed_df_office_only = []

for i in aed_df_office:
    if i[-1] == "터" and i[-4] == "주":
        list_aed_df_office_only.append(i)
        
list_aed_df_office_only.sort()

print(len(list_aed_df_office_only))
print(list_aed_df_office_only)
```

```python
public_office_df = pd.read_csv('data/서울시 공공시설 .csv', encoding='euc-kr')
list_public_office_df_dong_only = []
list_public_office_df_dong_a = list(public_office_df["개방시설명"])
list_public_office_df_dong_a_1 = []

for i in list_public_office_df_dong_a:
    if i[-1] == "터" and i[-4] == "주":
        list_public_office_df_dong_a_1.append(i)
        
list_public_office_df_dong_b = list(public_office_df["개방장소명"])
list_public_office_df_dong_union = list_public_office_df_dong_a_1 + list_public_office_df_dong_b
list_public_office_df_dong_union = set(list_public_office_df_dong_union)
# print(list_public_office_df_dong_union)


for i in list_public_office_df_dong_union:
    if i[-1] == "터" and i[-4] == "주":
        list_public_office_df_dong_only.append(i)
        #set_public_office_df_dong_only.append(i)
        
set_public_office_df_dong_only = set(list_public_office_df_dong_only)
set_public_office_df_dong_only = list(set_public_office_df_dong_only)
set_public_office_df_dong_only.sort()
#print(len(list_public_office_df_dong_only))        
#print(sorted(list_public_office_df_dong_only))

print(len(set_public_office_df_dong_only))
sample_li = set_public_office_df_dong_only
set_public_office_df_dong_only = [i.replace(' ', '') for i in sample_li]
print(set_public_office_df_dong_only)
```

```python
list_no_aed_office = []

for i in set_public_office_df_dong_only:
    if i not in list_aed_df_office_only:
        list_no_aed_office.append(i)

print(len(list_no_aed_office))        
print(sorted(list_no_aed_office))
```
