### 학습용/검증용 데이터 분리

```python
# train set, test set으로 분리
#import sklearn.model_selection as ms
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size = 0.3,\
                 random_state = 42)

# X_train, X_test, y_train, y_test
```

### 모델 구축

```python
# 1) 분석을 수행하기 위한 추정자(estimator) 생성
# 2) 학습 수행: 생성된 객체에 대해 fit() 메소드 수행
# 3) 예측 결과 도출: 생성된 객체 또는 학습이 수행된 분석 결과에 대해 predict()메소드 수행
```

```python
from sklearn.tree import DecisionTreeClassifier

# 1) 분석을 수행하기 위한 추정자(estimtor) 생성
cls = DecisionTreeClassifier()

# 2) 학습 수행: 생성된 객체에 대해 fit() 메서드 수행
cls.fit(X_train, y_train)

# 3) 예측 결과 수행
y_pred = cls.predict(X_test)
y_pred
```

### 모델 평가

```python
# 생성된 객체 또는 학습이 수행된 분석 결과에 대해 적합한 성능 평가 지표 도출
# from sklearn.metrics import ****
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```
